```python
import pandas as pd

df = pd.read_csv('nifty_500.csv')
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10, 6))

# Sort by 'Value (Indian Rupee)' which represents total traded value
top10 = df.sort_values(by='Value (Indian Rupee)', ascending=False).head(10)

sns.barplot(x='Value (Indian Rupee)', y='Company Name', data=top10, palette='viridis')
plt.title('ðŸ’° Top 10 Companies by Trading Value')
plt.xlabel('Trading Value (â‚¹)')
plt.ylabel('Company')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

top_industries = df['Industry'].value_counts().index[:10]

sns.countplot(y='Industry', data=df, order=top_industries, palette='mako')
plt.title('ðŸ­ Top 10 Industries by Company Count')
plt.xlabel('Number of Companies')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x='Last Traded Price',
    y='Share Volume',
    hue='Industry',
    size='Share Volume',
    alpha=0.6,
    legend=False
)
plt.title('ðŸ“ˆ Share Volume vs Last Traded Price by Industry')
plt.xlabel('Last Traded Price (â‚¹)')
plt.ylabel('Share Volume')
plt.tight_layout()
plt.show()
```
